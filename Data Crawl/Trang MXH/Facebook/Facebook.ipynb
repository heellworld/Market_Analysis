```python
def parse_time(text):
    # Sá»­ dá»¥ng regular expressions Ä‘á»ƒ tÃ¬m sá»‘ vÃ  Ä‘Æ¡n vá»‹ thá»i gian
    match = re.search(r'(\d+)\s*(phÃºt|giá»|ngÃ y)', text)
    if not match:
        return None
    
    amount, unit = match.groups()
    amount = int(amount)
    
    if 'phÃºt' in unit:
        return datetime.timedelta(minutes=amount)
    elif 'giá»' in unit:
        return datetime.timedelta(hours=amount)
    elif 'ngÃ y' in unit:
        return datetime.timedelta(days=amount)
    else:
        return None
```
