```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import TimeoutException,NoSuchElementException  # Import TimeoutException
import pandas as pd
import undetected_chromedriver as uc
import numpy as np
import os
from datetime import datetime
import random
from time import sleep
import time
```

```python
df = pd.read_csv(r"D:\projects\NCKH\Data Crawl\Trang web bán hàng\FPT Shop\Link_product_fpt.csv")
list_links = df['links'].to_list()
list_links
```

## Get Infooo

```python
def get_comment(driver, link):
    products = []  # Danh sách để lưu trữ các comment
    driver.get(link)
    sleep(5)  # Đợi trang tải và cookies có hiệu lực

    print(link)

    try:
        # Lấy ngày hiện tại
        current_date = datetime.now().strftime("%Y-%m-%d")
        print(current_date)
        
        # Sử dụng hàm safe_find_text để tìm và trả về văn bản của các phần tử, hoặc None nếu không tìm thấy
        product_title = safe_find_text(driver, ".st-name")
        current_price = safe_find_text(driver, ".st-price-sub")
        discounted_price = safe_find_text(driver, ".st-price-main")
        discount = safe_find_text(driver, "#root > main > div > div.l-pd-header > div:nth-child(2) > div.l-pd-row.clearfix > div.l-pd-right > div.st-price.npi-special > div.st-price__left.flex.flex-column.npi-special-price.active.false > div > div.flex.npi-special-price-sub > span")
        rating_averrage = safe_find_text(driver, "#root-review > div > div > div > div.card-body > div.user-rate__box > div > div:nth-child(1) > div > div.f-s-ui-44.text-primary.f-w-500.m-t-4")
        rating_count = safe_find_text(driver, "#re-rate")

        product_info = {
            'Platform': 'FPT',
            'URL': link,
            'Date Crawled': current_date,
            'Product Title': product_title,
            'Current Price': current_price,
            'Discounted Price': discounted_price,
            'Discount': discount,
            'Rating Average': rating_averrage,
            'Rating Count': rating_count,
        }
        
        products.append(product_info)
    except TimeoutException:
        print("Timed out waiting for page to load")
    finally:
        return products

def safe_find_text(driver, selector, by=By.CSS_SELECTOR):
    try:
        element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((by, selector)))
        return element.text
    except (NoSuchElementException, TimeoutException):
        return None
```

```python
# Khởi tạo undetected_chromedriver với các tuỳ chỉnh
options = uc.ChromeOptions()
options.add_argument("--disable-extensions")
options.add_argument("--start-maximized")
driver = uc.Chrome(options=options)
driver.implicitly_wait(10)

products = []
for link in list_links:
    products.extend(get_comment(driver, link))

# Sau khi đã lấy xong dữ liệu, đóng trình duyệt
driver.quit()

# Tạo DataFrame từ danh sách comments và xuất ra CSV
comments_df = pd.DataFrame(products)
```

```python
comments_df.to_csv('product_smartphone_FPT.csv', mode='a', header=not os.path.exists('product_smartphone_FPT.csv'), index=False, encoding='utf-8-sig')
```

## Get Comment

```python
def get_commentt(driver, link):
    comments_list = []  # Danh sách để lưu trữ các comment
    driver.get(link)
    sleep(10)  # Đợi trang tải và cookies có hiệu lực

    print(link)

    try:
        comment_overview = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CSS_SELECTOR, '#root-review > div > div > div > div.card-body > div.user-content > div.user-wrapper')))
        driver.execute_script("arguments[0].scrollIntoView(true);", comment_overview)
        sleep(1)
    except Exception as e:
        print(f"Không thể tìm thấy hoặc scroll đến bình luận. Lỗi: {e}")
        return comments_list  # Trả về danh sách trống nếu không tìm thấy phần tổng quan bình luận

    while True:
        try:
            customer_names = safe_find_elements(driver, '#root-review > div > div > div > div.card-body > div.user-content > div.user-wrapper > div:nth-child(1) > div > div.avatar-info > div.avatar-name > div')
            time_comments = safe_find_elements(driver, '#root-review > div > div > div > div.card-body > div.user-content > div.user-wrapper > div:nth-child(1) > div > div.avatar-info > div.avatar-time > div.text.text-grayscale')
            contents = safe_find_elements(driver, "#root-review > div > div > div > div.card-body > div.user-content > div.user-wrapper > div:nth-child(1) > div > div.avatar-info > div.avatar-para > div")

            for i in range(len(customer_names)):
                comment_info = {
                    'Platform': 'Shopee',
                    'URL': link,
                    'Customer Name': customer_names[i],
                    'Time Comment': time_comments[i] if i < len(time_comments) else None,
                    'Content': contents[i] if i < len(contents) else None,
                }
                comments_list.append(comment_info)

            try:
                # Tìm nút Next page dựa trên class của nó
                next_page = driver.find_element(By.XPATH, '//i[contains(@class, "cm-ic-angle-right")]')
                if next_page:
                    driver.execute_script("arguments[0].click();", next_page)
                    sleep(5)  # Đợi trang mới tải xong
                else:
                    break  # Nếu không có nút Next page, thoát khỏi vòng lặp
            except NoSuchElementException:
                break  # Thoát khỏi vòng lặp nếu không tìm thấy nút Next page
        except TimeoutException:
            print("Không tìm thấy bình luận cho link này. Chuyển sang link tiếp theo.")
            break

    return comments_list

def safe_find_elements(driver, selector, by=By.CSS_SELECTOR):
    try:
        elements = WebDriverWait(driver, 5).until(EC.presence_of_all_elements_located((by, selector)))
        return [element.text for element in elements]
    except (NoSuchElementException, TimeoutException):
        return []
```

```python
# Khởi tạo undetected_chromedriver với các tuỳ chỉnh
options = uc.ChromeOptions()
options.add_argument("--disable-extensions")
options.add_argument("--start-maximized")
driver = uc.Chrome(options=options)
driver.implicitly_wait(10)

comments = []
for link in list_links:
    comments.extend(get_commentt(driver, link))

# Sau khi đã lấy xong dữ liệu, đóng trình duyệt
driver.quit()

# Tạo DataFrame từ danh sách comments và xuất ra CSV
df = pd.DataFrame(comments)
df.to_csv('comments_fpt_28-3.csv', mode='a', header=not os.path.exists('comments_fpt_28-3.csv'), index=False, encoding='utf-8-sig')
print(df)
```
